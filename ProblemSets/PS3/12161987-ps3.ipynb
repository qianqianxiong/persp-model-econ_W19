```python
import numpy as np
w_lb = 1e-2
w_ub = 1.0
N = 100
w_vec = np.linspace(w_lb, w_ub, N)
print(w_vec)
```

```python
#Problem 5.10
```

```python
#choose a utility function u(c)=log(c)
def utility(c):
    util = np.log(c)
    return util
```

```python
beta = 0.9
```

```python
# initial guess for value function
v_prime = np.zeros(N).reshape(N,1)
v_init = utility(w_vec)
c_mat = (np.tile(w_vec.reshape((N, 1)), (1, N)) - 
         np.tile(w_vec.reshape((1, N)), (N, 1)))
c_pos = c_mat > 0
c_mat[~c_pos] = 1e-7
u_mat = utility(c_mat)
```

```python
print(u_mat)
```

```python
V = u_mat + beta * v_prime
v_new = V.max(axis = 1)
array = np.argmax(u_mat + beta * v_prime,axis = 1)
w_vec[array]
```

```python
plt.plot(w_vec,w_vec[array])
plt.title("Policy function")
plt.show()
```

```python
plt.plot(v_new[1:])
plt.title("Value function")
plt.show()
```

```python
N = 100
c_mat = (np.tile(w_vec.reshape((N,1)),(1,N)))-(np.tile(w_vec.reshape((1,N)),(N,1)))
c_pos = c_mat > 0
c_mat[~c_pos]=1e-9
u_mat = utility(c_mat)
```

```python
dist = 10.0
v_iter = 0
maxit = 500
tool = 1e-9
v_init = utility(w_vec)
```

```python
while dist > tool and v_iter < maxit:
    v_iter += 1
    v_prime = np.tile(v_init.reshape((1,N)),(N,1))
    v_prime[~c_pos] = -9e+4
    V = u_mat + beta * v_prime
    v_new = V.max(axis = 1)
    array = np.argmax(V,axis=1)
    dist = ((v_new-v_init) ** 2).sum()
    print("Iter=",v_iter,",distance=",dist)
    v_init=v_new
print("The convergence takes",v_iter,"iterations")
print("Is v_new equal to v_init ?",np.array_equal(v_init,v_new))
```

```python
#Problem 5.17
```

```python
c_mat = np.tile(w_vec.reshape((N,1)), (1,N)) - np.tile(w_vec.reshape((1,N)), (N,1))
c_pos = c_mat > 0
c_mat[~c_pos] = 1e-9
u_mat = utility(c_mat)
C = np.array([u_mat * e for e in ep])
```

```python
v_init = np.zeros((N,M))
ev = v_init @ pdf.reshape((M,1)) 
ev_mat = np.tile(ev.reshape((1,N)), (N,1))
ev_mat[~c_pos] = -9e+4
ev_C = np.array([ev_mat for e in range(M)])
```

```python
vt = C + beta * ev_C
v_new = np.zeros((N,M))
w_new = np.zeros((N,M))
for i in range(N):
    V = vt[:, i, :]
    v_new[i] = V.max(axis = 1)
    array = np.argmax(V, axis = 1)
    w_new[i] = w_vec[array]
```

```python
plt.plot(w_vec,np.average(w_new,axis=1))
plt.title("Policy function")
plt.show()
```

```python
plt.plot(np.average(v_new,axis = 1))
plt.title("Value function")
plt.show()
```
